# 9663 N-Queen

퀸의 공격 범위는 8방위로 모든 칸

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

N = int(input())
answer = 0
all = set((x, y) for x in range(N) for y in range(N))

def check_safe(i, j, banned, depth):
    global answer

    if depth >= N:
        answer += 1
        return

    for k in range(N):
        banned.add((i, k)) # 같은 행
        banned.add((k, j)) # 같은 열

    # 대각선
    dx = [-1, 1, -1, 1]
    dy = [1, -1, -1, 1]

    for k in range(4):
        ni, nj = i, j
        while True:
            ni += dx[k]
            nj += dy[k]
            if not (0 <= ni < N and 0 <= nj < N):
                break
            banned.add((ni, nj))

    for x, y in all - banned:
        # return을 쓰면 거기서 끝나 버림!!!
        banned.add((x, y))
        check_safe(x, y, banned.copy(), depth + 1)

start = set()
for i in range(N): # 행
    for j in range(N): # 열
        check_safe(i, j, start.copy(), 1)
        start.add((i, j))

print(answer)
```

# 1987 알파벳(보류)

```python
R, C = map(int, input().split())
# import sys
# input = lambda: sys.stdin.readline().rstrip()

grid = []
answer = 0

for _ in range(R):
    grid.append(list(input()))

def dfs(x, y, depth, alph):
    global answer
    if 0 <= x < R and 0 <= y < C:
        if grid[x][y] not in alph:
            answer = max(answer, depth)
            dx = [-1, 0, 1, 0]
            dy = [0, -1, 0, 1]

            for i in range(4):
                nx, ny = x + dx[i], y + dy[i]
                dfs(nx, ny, depth + 1, alph | {grid[x][y]})

dfs(0, 0, 1, set())
print(answer)
```

# 1043 거짓말

```python
from collections import deque

# import sys
# input = lambda: sys.stdin.readline().rstrip()

N, M = map(int, input().split())
true_people = set(map(int, input().split()[1:]))
people_party = dict({k: [] for k in range(1, N + 1)})
truth = [False] * (M + 1)

party = [0 for _ in range(M + 1)]
queue = deque([])

for i in range(1, M + 1):
    party_list = list(map(int, input().split()[1:]))
    party[i] = party_list
    for p in party_list:
        people_party[p].append(i)
        if p in true_people:
            queue.append(i)
            truth[i] = True

while queue:
    # 진실을 아는 사람이 1명이라도 있는 파티
    curr = queue.popleft()

    # 거기 있는 사람들도 진실을 알게 됨
    for other in party[curr]:
        for next in people_party[other]:
            if not truth[next]:
                truth[next] = True
                queue.append(next)

print(M - sum(truth))
```

# 9465 스티커

```python
T = int(input())

def stickers(N):
    grid = []
    for _ in range(2):
        grid.append(list(map(int, input().split())))
    dp = [[0] * 3 for _ in range(N)]

    # 0: 위 선택 1: 아래선택 2: 미선택
    dp[0][0] = grid[0][0]
    dp[0][1] = grid[1][0]
    dp[0][2] = 0


    for i in range(1, N):
        dp[i][0] = max(dp[i-1][1], dp[i-1][2]) + grid[0][i]
        dp[i][1] = max(dp[i-1][0], dp[i-1][2]) + grid[1][i]
        dp[i][2] = max(dp[i-1][0], dp[i-1][1], dp[i-1][2])

    return max(dp[N-1])

for _ in range(T):
    N = int(input())
    print(stickers(N))
```

# 10830 행렬제곱

```python
# import sys
# input = lambda: sys.stdin.readline().rstrip()

N, B = map(int, input().split())
A = []

for _ in range(N):
    A.append(list(map(int, input().split())))

dp = dict()

def matmul(X, Y):
    Z = [[0] * N for _ in range(N)]

    for i in range(N):
        for j in range(N):
            result = 0
            for k in range(N):
                result += X[i][k] * Y[k][j]
            Z[i][j] = result % 1000

    return Z

def matpower(A, B):
    if B not in dp:
        if B == 1:
            dp[B] = A
        elif B % 2 == 0:
            C = matpower(A, B // 2)
            dp[B] = matmul(C, C)
        else:
            C = matpower(A, B // 2)
            D = matpower(A, B // 2 + 1)
            dp[B] = matmul(C, D)
    return dp[B]


result = matpower(A, B)
for i in range(N):
    for j in range(N):
        print(result[i][j] % 1000, end=" ")
    print()

# 맨 처음에 1000이 들어올 수 있음.
```

```python
#import sys
#input = lambda: sys.stdin.readline().rstrip()

A, B, C = map(int, input().split())
dp = dict()
dp[1] = A % C

def cal(A, B, C):
    if B not in dp:
        if B % 2 == 0:
            dp[B] = (cal(A, B // 2, C) ** 2) % C
        else:
            dp[B] = (cal(A, B // 2, C) * cal(A, B // 2 + 1, C)) % C
    return dp[B]


print(cal(A, B, C))
```

# 1167 트리의 지름

* 자식 노드가 3개 이상일 수 있다.
* 간선의 거리가 음수일 수 있다.

```python
import sys
sys.setrecursionlimit(10**6)
# input = lambda: sys.stdin.readline().rstrip()

V = int(input())
tree = [[] for _ in range(V + 1)]
visited = [False] * (V + 1)

for _ in range(V):
    input_list = list(map(int, input().split()))
    curr = input_list[0]
    for i in range(1, len(input_list) - 1, 2):
        tree[curr].append((input_list[i+1], input_list[i]))

tree_max = 0

def check_cost(node):
    global tree_max
    visited[node] = True
    results = []
    for cost, next in tree[node]:
        if not visited[next]:
            results.append(cost + check_cost(next))

    if not results:
        return 0

    results.sort(reverse=True)
    tree_max = max(tree_max, *results, sum(results[:2]))
    return max(results)

check_cost(1)
print(tree_max)
```

```python
# 모범답안: DFS 하면서 깊이 정보를 계속 저장해 두기.

import sys
sys.setrecursionlimit(10**6)
input = sys.stdin.readline

V = int(input())
tree = [[] for _ in range(V + 1)]

# 트리 입력 받기
for _ in range(V):
    input_list = list(map(int, input().split()))
    curr = input_list[0]
    for i in range(1, len(input_list) - 1, 2):
        next_node, cost = input_list[i], input_list[i + 1]
        tree[curr].append((next_node, cost))
        tree[next_node].append((curr, cost))  # 양방향 연결

# 가장 먼 노드를 찾는 DFS 함수
def dfs(node, dist):
    global farthest_node, max_dist
    visited[node] = True

    if dist > max_dist:
        max_dist = dist
        farthest_node = node  # 가장 먼 노드 갱신

    for next_node, cost in tree[node]:
        if not visited[next_node]:
            dfs(next_node, dist + cost)

# 1. 임의의 점 (1)에서 가장 먼 노드 찾기
visited = [False] * (V + 1)
farthest_node = 0
max_dist = 0
dfs(1, 0)

# 2. 찾은 노드에서 다시 DFS 실행하여 트리의 지름 구하기
visited = [False] * (V + 1)
max_dist = 0
dfs(farthest_node, 0)

# 결과 출력 (트리의 지름)
print(max_dist)
```

# 14938 서강그라운드

```python
import heapq
# import sys
# input = lambda: sys.stdin.readline().rstrip()

N, M, R = map(int, input().split())
t_list = [0] + list(map(int, input().split()))
grid = [[] for _ in range(N + 1)]

for _ in range(R):
    a, b, l = map(int, input().split())
    grid[a].append((l, b))
    grid[b].append((l, a))

def dijkstra(start):
    dist = [int(1e9)] * (N + 1)
    queue = []
    heapq.heappush(queue, (0, start))
    dist[start] = 0

    while queue:
        cost, curr = heapq.heappop(queue)
        if cost > dist[curr]:
            continue
        for nextcost, next in grid[curr]:
            finalcost = nextcost + dist[curr]
            if finalcost < dist[next]:
                dist[next] = finalcost
                heapq.heappush(queue, (finalcost, next))

    answer = 0
    for i in range(1, N + 1):
        if dist[i] <= M:
            answer += t_list[i]

    return answer

final_answer = max(dijkstra(i) for i in range(1, N + 1))
print(final_answer)
```

# 11779 최소비용 구하기 2

```python
import heapq
N = int(input())
M = int(input())
graph = [[] for _ in range(N + 1)]

for _ in range(M):
    s, e, cost = map(int, input().split())
    graph[s].append((cost, e))

START, DEST = map(int, input().split())

def dijkstra(start, end):
    queue = []
    visited = [int(1e9)] * (N + 1)
    route = [None] * (N + 1)
    heapq.heappush(queue, (0, start))
    visited[start] = 0
    route[start] = [start]

    while queue:
        cost, curr = heapq.heappop(queue)
        if cost > visited[curr]:
            continue
        for next_cost, next in graph[curr]:
            final_cost = next_cost + visited[curr]
            if final_cost < visited[next]:
                visited[next] = final_cost
                route[next] = route[curr] + [next]
                heapq.heappush(queue, (final_cost, next))

    return visited[end], route[end]

answer1, answer3 = dijkstra(START, DEST)
print(answer1)
print(len(answer3))

for a in answer3:
    print(a, end=" ")
```

# 1238 파티

```python
import heapq
# import sys
# input = lambda: sys.stdin.readline().rstrip()

N, M, X = map(int, input().split())
graph = [[] for _ in range(N + 1)]
INF = int(1e9)

for _ in range(M):
    start, end, cost = map(int, input().split())
    graph[start].append((cost, end))

def check_cost(start):
    queue = []
    distance = [INF] * (N + 1)
    heapq.heappush(queue, (0, start))
    distance[start] = 0

    while queue:
        cost, curr = heapq.heappop(queue)
        if cost > distance[curr]:
            continue
        for next_cost, next in graph[curr]:
            final_cost = distance[curr] + next_cost
            if final_cost < distance[next]:
                distance[next] = final_cost
                heapq.heappush(queue, (final_cost, next))
    return distance

costs = [None] + [check_cost(i) for i in range(1, N + 1)]

answer = max(costs[i][X] + costs[X][i] for i in range(1, N + 1))
print(answer)
```

# 1504 특정한 최단 경로

```python
# import sys
# input = lambda: sys.stdin.readline().rstrip()
import heapq

N, E = map(int, input().split())
INF = int(1e9)
grid = [[] for _ in range(N + 1)]

for _ in range(E):
    a, b, c = map(int, input().split())
    grid[a].append((b, c))
    grid[b].append((a, c))

def djikstra(start):
    dist = [INF] * (N + 1)
    queue = []
    heapq.heappush(queue, (0, start))
    dist[start] = 0

    while queue:
        cost, curr = heapq.heappop(queue)
        if cost > dist[curr]:
            continue
        for next, next_cost in grid[curr]:
            final_cost = dist[curr] + next_cost
            if final_cost < dist[next]:
                heapq.heappush(queue, (final_cost, next))
                dist[next] = final_cost

    return dist

v1, v2 = map(int, input().split())

start_dist = djikstra(1)
v1_dist = djikstra(v1)
v2_dist = djikstra(v2)

route1 = start_dist[v1] + v1_dist[v2] + v2_dist[N]
route2 = start_dist[v2] + v2_dist[v1] + v1_dist[N]

if route1 >= INF and route2 >= INF:
    print(-1)
else:
    print(min(route1, route2))
```

# 13549 숨바꼭질 3

```python
import heapq
# import sys
# input = lambda: sys.stdin.readline().rstrip()

INF = int(1e9)
N, K = map(int, input().split())
distance = [INF] * (100001)
distance[N] = 0

queue = []
heapq.heappush(queue, (0, N))

while queue:
    time, curr = heapq.heappop(queue)

    if distance[curr] < time:
        continue

    for next_cost, next in [(time + 1, curr - 1), (time + 1, curr + 1), (time, curr * 2)]:
        if 0 <= next <= 100000 and next_cost < distance[next]:
            distance[next] = next_cost
            heapq.heappush(queue, (next_cost, next))

print(distance[K])
```

# 1916 최소비용 구하기

```python
# 다익스트라 복습겸
import heapq
# import sys
# input = lambda: sys.stdin.readline().rstrip()

INF = int(1e9)
N = int(input())
M = int(input())
graph = [[] for _ in range(N + 1)]
distance = [INF] * (N + 1)

for _ in range(M):
    start, end, cost = map(int, input().split())
    graph[start].append((end, cost))

q_st, q_en = map(int, input().split())
distance[q_st] = 0
queue = []
heapq.heappush(queue, (0, q_st))

while queue:
    dist, curr = heapq.heappop(queue)
    if distance[curr] < dist:
        continue
    for next, next_dist in graph[curr]:
        cost = dist + next_dist
        if cost < distance[next]:
            distance[next] = cost
            heapq.heappush(queue, (cost, next))

print(distance[q_en])
```

# 16953 A -> B

```python
from collections import deque
# import sys
# input = lambda: sys.stdin.readline().rstrip()

A, B = map(int, input().split())

queue = deque([(A, 1)])
visited = set([A])

def bfs():
    while queue:
        curr, dist = queue.popleft()
        for next in (curr * 2, int(str(curr) + "1")):
            if next == B:
                return dist + 1
            elif next < B and next not in visited:
                queue.append((next, dist + 1))
                visited.add(next)

    return -1


print(bfs())
```
