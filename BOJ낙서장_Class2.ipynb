# 4949 균형잡힌 세상

```python
par_match = {")": "(", "]": "["}

while True:
    check_line = input()
    if check_line == ".":
        break
    else:
        stack = list()
        for lett in check_line:
            if lett == "(" or lett == "[":
                stack.append(lett)
            elif lett == ")" or lett == "]":
                if stack and stack[-1] == par_match[lett]:
                    stack.pop()
                else:
                    print("no")
                    break

        else:
            if stack:
                print("no")
            else:
                print("yes")
```

# 2775 부녀회장이 될테야

```python
t = int(input())
apart = dict()

def people(floor, room):
    if (floor, room) in apart:
        return apart[(floor, room)]
    elif floor == 0: # 종료조건
        apart[(floor, room)] = room
        return room
    else:
        lower_sum = 0
        for idx in range(1, room + 1):
            lower_sum += people(floor - 1, idx)
        apart[(floor, room)] = lower_sum
        return lower_sum

for _ in range(t):
    k = int(input())
    n = int(input())
    print(people(k, n))
```

# 28702 FizzBuzz

```python
def f_out(i):
    if i % 3 == 0 and i % 5 == 0:
        return "FizzBuzz"
    elif i % 3 == 0:
        return "Fizz"
    elif i % 5 == 0:
        return "Buzz"
    else:
        return i

first = input()
second = input()
third = input()

if third == "FizzBuzz":
    print(f_out(int(second) + 2))
elif third == "Fizz":
    if second == "Buzz":
        print(f_out(int(first) + 3))
    else:
        print(f_out(int(second) + 2))
elif third == "Buzz":
    if second == "Fizz":
        print(f_out(int(first) + 3))
    else:
        print(f_out(int(second) + 2))
else:
    print(f_out(int(third) + 1))
```

# 1676 팩토리얼 0의 개수

```python
import math
n = int(input())
n_string = str(math.factorial(n))
answer = 0

for digit in n_string[::-1]:
    if digit != "0":
        print(answer)
        break
    else:
        answer += 1
```

```python
import sys

n = int(input())

# Python의 round 방식 우회...
def custom_round(num):
    return int(num + 0.5)

if n == 0:
    print(0)
else:
    remove = custom_round(n * 0.15)
    score_list = []
    for _ in range(n):
        score_list.append(int(sys.stdin.readline().rstrip()))
    score_list.sort()
    score_list = score_list[remove: len(score_list) - remove]
    final_score = custom_round(sum(score_list) / len(score_list))
    print(final_score)
```

# 11866 요세푸스 문제 0

```python
n, k = map(int, input().split())
people = [i for i in range(1, n + 1)]
answer = []

while people:
    out = (k - 1) % len(people)
    answer.append(people[out])
    people = people[out + 1:] + people[:out]

print("<", end="")
print(*answer, sep=', ', end="")
print(">")
```

# ★ 1018 체스판 다시 칠하기

* 행렬 안에서 굳이 다른 행렬을 만들기보단, 별개의 인덱스를 지정하는 게 편하다.

```python
n, m = map(int, input().split())
grid = []

for _ in range(n):
    grid.append(list(input()))

answer = 1e9

for r in range(n - 8 + 1):
    for c in range(m - 8 + 1):
        count = 0
        for dr in range(8):
            for dc in range(8):
                if (dr + dc) % 2 == 0 and grid[r + dr][c + dc] == "W":
                    count += 1
                elif (dr + dc) % 2 == 1 and grid[r + dr][c + dc] == "B":
                    count += 1
        answer = min(answer, min(count, 64 - count))

print(answer)
```

# 9012 괄호

```python
n = int(input())

for _ in range(n):
    stack = []
    check_str = input()
    for lett in check_str:
        if lett == "(":
            stack.append("(")
        elif lett == ")":
            if stack:
                stack.pop()
            else:
                print("NO")
                break
    else:
        if stack:
            print("NO")
        else:
            print("YES")
```

# 10845 큐

```python
from collections import deque
queue = deque()
n = int(input())

for _ in range(n):
    command = input()
    if command.startswith('push'):
        input_data = int(command.split()[1])
        queue.append(input_data)
    elif command == "pop":
        if not queue:
            print(-1)
        else:
            print(queue.popleft())
    elif command == "size":
        print(len(queue))
    elif command == "empty":
        if queue:
            print(0)
        else:
            print(1)
    elif command == "front":
        if queue:
            print(queue[0])
        else:
            print(-1)
    elif command == "back":
        if queue:
            print(queue[-1])
        else:
            print(-1)
```

# 10828 스택

```python
n = int(input())
stack = []

for _ in range(n):
    command = input()
    if command.startswith('push'):
        input_data = int(command.split()[1])
        stack.append(input_data)
    elif command == "pop":
        if not stack:
            print(-1)
        else:
            print(stack.pop())
    elif command == "size":
        print(len(stack))
    elif command == "empty":
        if stack:
            print(0)
        else:
            print(1)
    elif command == "top":
        if stack:
            print(stack[-1])
        else:
            print(-1)
```
