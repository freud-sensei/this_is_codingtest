# Q31 금광

```python
t = int(input())


for _ in range(t):
    n, m = map(int, input().split())
    input_list = list(map(int, input().split()))
    grid = [input_list[m * idx: m * (idx + 1)] for idx in range(n)]

    dr = [-1, 0, 1]

    for col in range(1, m):
        for row in range(n):
            check = []
            for i in range(3):
                if 0 <= row + dr[i] < n:
                    check.append(grid[row + dr[i]][col-1])
            grid[row][col] += max(check)

    print(max(grid[row][m - 1] for row in range(n)))

```

$d_{ij} = d_{ij} + max(d_{i-1, j-1}, d_{i, j-1}, d_{i+1, j-1})$ 를 구현한다.


```python
# 모범답안

t = int(input())


for _ in range(t):
    n, m = map(int, input().split())
    input_list = list(map(int, input().split()))
    grid = [input_list[m * idx: m * (idx + 1)] for idx in range(n)]

    dr = [-1, 0, 1]

    for col in range(1, m):
        for row in range(n):
            check = []
            if row != 0:
                check.append(grid[row - 1][col - 1])
            check.append(grid[row][col - 1])
            if row != n - 1:
                check.append(grid[row + 1][col - 1])
            grid[row][col] += max(check)

    print(max(grid[row][m - 1] for row in range(n)))

```

# Q32 정수 삼각형

```python
n = int(input())
tri = []
for _ in range(n):
    tri.append(list(map(int, input().split())))

for i in range(1, n):
    for j in range(i + 1):
        check = []
        if j != 0:
            check.append(tri[i-1][j-1])
        if j != i:
            check.append(tri[i-1][j])
        tri[i][j] += max(check)

print(max(tri[n - 1][j] for j in range(n)))
```

$a_{ij} = a_{ij} + max(a_{(i-1, j-1)}, a_{(i-1, j)})$ 를 구현한다.

# Q34 병사 배치하기

```python
import sys
sys.setrecursionlimit(10 ** 6)
input = sys.stdin.readline

n = int(input())
power = list(map(int, input().split()))
power = [1e9] + power
memo = dict()

def check(i, smallest):
    if i == n + 1:
        return 0
    if (i, smallest) not in memo:
        if power[i] < smallest:
            take = check(i + 1, power[i]) + 1
        else:
            take = 0
        out = check(i + 1, smallest)
        memo[(i, smallest)] = max(take, out)
    return memo[(i, smallest)]

print(n - check(1, power[0]))
```

점화식은 잘 생각이 안 나서 재귀로 풀어 봤는데, 점화식으로도 가능..? 점화식 없이도 메모이제이션 쓰니까 어째 정답이 나오긴 했지만

* **가장 증가하는 부분 수열(LIS)**이라고도 불리는 다이나믹 프로그래밍 문제.
* `D[i]`: `array[i]`를 **마지막 원소로 가지는** 부분 수열의 최대 길이`로 두면,

모든 `0 <= j < i`에 대하여, `D[i] = max(D[i], D[j] + 1) if array[j] < array[i]` 를 구현한다.

본 문제는 감소하는 케이스이므로,

모든 `0 <= j < i`에 대하여, `D[i] = max(D[i], D[j] + 1) if array[j] > array[i]` 를 구현한다.

```python
n = int(input())
power = list(map(int, input().split()))
dp = [1] * n

for i in range(1, n):
    for j in range(i):
        if power[j] > power[i]:
            dp[i] = max(dp[i], dp[j] + 1)

print(n - max(dp))
```

# Q35 못생긴 수

```python
n = int(input())
ugly_list = [0]

def ugly(x):
    if len(ugly_list) - 1 < x:
        if x == 1:
            ugly_list.append(True)
        elif x % 2 == 0:
            ugly_list.append(ugly_list[x // 2])
        elif x % 3 == 0:
            ugly_list.append(ugly_list[x // 3])
        elif x % 5 == 0:
            ugly_list.append(ugly_list[x // 5])
        else:
            ugly_list.append(False)

    return ugly_list[x]

i = 0
count = 0

while count < n:
    i += 1
    if ugly(i):
        count += 1

print(i)
```

* 당연히 나눗셈일 거라고 생각했는데, 이 분은 곱셈으로 푸심??
* 못생긴 수에 2, 3, 5를 곱한 수도 못생긴 수
* 순서가 섞이는 것을 막기 위해, 매번 가능한 곱셈 결과 중 가장 작은 수만을 선택해야 함

```python
n = int(input())
ugly = [0] * n # 못생긴 수 저장
ugly[0] = 1

i2 = i3 = i5 = 0
next2, next3, next5 = 2, 3, 5

# n번째 못생긴 수 찾기
for num in range(1, n):
    # 가능한 곱셈 결과 중 가장 작은 수만 선택
    ugly[num] = min(next2, next3, next5)
    if ugly[num] == next2:
        i2 += 1
        next2 = ugly[i2] * 2
    if ugly[num] == next3:
        i3 += 1
        next3 = ugly[i3] * 3
    if ugly[num] == next5:
        i5 += 1
        next5 = ugly[i5] * 5

print(ugly[n - 1])
```
