# Q1 모험가 길드

```python
# 나의 답안

n = int(input())
exp_list = list(map(int, input().split()))
exp_list.sort()

answer = 0 # 문제의 정답 - 총 만들 수 있는 길드 수
quota = 0 # 현재 길드가 출발하려면 필요한 인원 수
current = 0 # 현재 길드의 인원 수
idx = 0

while idx < n:
  # 길드에 새 사람이 들어왓어요
  quota = exp_list[idx] # 필요한 인원 수가 늘 수도 있죠
  current += 1 # 인원이 한 명 늘죠

  if quota == current: # 길드가 출발해요
    answer += 1 # 길드 1개 추가
    quota, current = 0, 0 # reset

  idx += 1 # 다음 사람으로 넘어가요

print(answer)
```

* **오름차순**으로 정렬을 수행한다.
* 현재 길드의 인원 수가 현재 확인하고 있는 공포도보다 크거나 같으면, 길드를 결성한다.

```python
# 모범 답안 기반 수정

n = int(input())
exp_list = list(map(int, input().split()))
exp_list.sort()

answer = 0 # 문제의 정답 - 총 만들 수 있는 길드 수
current = 0 # 현재 길드의 인원 수

for e in exp_list:
  # 길드에 새 사람이 들어왓어요
  current += 1 # 인원이 한 명 늘죠
  if e == current: # 길드가 출발해요
    answer += 1 # 길드 1개 추가
    current = 0

print(answer)
```

# Q2 곱하기 혹은 더하기

```python
# 나의 답안
numbers = list(map(int, input()))
answer = numbers[0]

for idx in range(1, len(numbers)):
  # 0과 1이 있는 경우 어쩔 수 없이 더하기
  if answer in (0, 1) or numbers[idx] in (0, 1):
    answer += numbers[idx]
  # 0과 1이 없는 경우 무조건 곱하기
  else:
    answer *= numbers[idx]

print(answer)
```

* 두 수에 대하여 연산을 수행할 때, 두 수 중 하나라도 1 이하이면 **더하기**, 모두 2 이상이면 **곱하기**.

```python
# 모범 답안 기반 수정
numbers = input()
answer = int(numbers[0])

for n in numbers[1:]:
  next = int(n)
  if answer <= 1 or int(n) <= 1:
    answer += next
  else:
    answer *= next

print(answer)
```

# Q3 문자열 뒤집기

https://www.acmicpc.net/problem/1439

```python
# 나의 답안
numbers = list(map(int, input()))
base = numbers[0]
answer = 0

for idx in range(1, len(numbers)):
  # 뒤집기 시작 시점만 카운트
  if numbers[idx] != base and numbers[idx - 1] == base:
    answer += 1

print(answer)
```

* 전부 0으로 바꾸는 경우와, 전부 1로 바꾸는 경우 중 더 적은 횟수를 가지는 경우를 계산한다.

```python
# 모범 답안 기반 수정
numbers = input()
count0 = 0 # 전부 0으로 바꾸는 경우
count1 = 0 # 전부 1로 바꾸는 경우

if numbers[0] == "1":
  count0 += 1
else:
  count1 += 1

for idx in range(len(numbers) - 1):
  if numbers[idx] != numbers[idx + 1]:
    if numbers[idx + 1] == "1":
      count0 += 1
    else:
      count1 += 1

print(min(count0, count1))
```

사견: 근데 내 위에 답도 인정이 된 걸 봐선, 맨 앞이 0인 경우엔 전부 0으로 바꾸는 게 덜 걸리고, 1인 경우엔 전부 1로 바꾸는 게 덜 걸리는 게 확실한 것 같다.

# Q4 만들 수 없는 금액 (★)

```python
n = int(input())
coins = list(map(int, input().split()))
coins.sort()
c_sum = 0

# 1원부터 시작하지 않는 경우 -> 답은 1원
if coins[0] > 1:
  answer = 1
else:
  for idx in range(len(coins)):
    c_sum += coins[idx]

    # 도중에 끝이 안 난 경우 -> 답은 동전 총합 + 1
    if idx == len(coins) - 1:
      answer = c_sum + 1
      break

    # 지금까지 동전 총합과, 다음으로 큰 동전 간 만들 수 없는 숫자가 생김
    if coins[idx + 1] - c_sum > 1:
      answer = c_sum + 1
      break

print(answer)
```

* 지금까진 `1`부터 `target-1`까지 만들 수 있었다고 가정한다.
* `target`도 만들 수 있는 지 확인한다. 만들 수 있으면 `target`의 값을 업데이트한다.
* 새로운 동전의 값이 `target`보다 작거나 같으면, `target`을 무조건 만들 수 있다.
* 하지만 새로운 동전의 값이 `target`보다 크면, `target-1`과 새로운 동전 사이 값의 `target`을 만들 방법이 없다.

예시 - 1 1 2 3 9
* `target=1`을 만족할 수 있는가? --> 우리에겐 1 동전이 있음.
* `target=2`를 만족할 수 있는가? --> 우리에겐 2번째 1 동전이 있음. (2까지의 모든 금액을 만들 수 있음)
* `target=3`을 만족할 수 있는가? --> 우리에겐 2 동전이 있음. (4까지의 모든 금액을 만들 수 있음)
* `target=5`를 만족할 수 있는가? --> 우리에겐 3 동전이 있음. (7까지의 모든 금액을 만들 수 있음.)
* `target=8`를 만족할 수 있는가? --> 8보다 큰 9 단위의 동전이 있지만, 8은 만들 수 없음.

```python
n = int(input())
coins = list(map(int, input().split()))
coins.sort()
target = 1
coin_sum = 0

for c in coins:
  if c > target:
    # target을 만들 수 없다면, 해당 target이 정답
    break
  else:
    # target을 만들 수 있다면, 이에 맞춰 갱신
    target += c

print(target)
```

# Q5 볼링공 고르기

```python
# 입력
n, m = map(int, input().split()) # 볼링공 개수, 최대무게
ball_list = list(map(int, input().split()))
ball_list.sort()
answer = 0
count = 1

for idx in range(1, n):
  # 무게가 달라졌을 때
  if ball_list[idx] != ball_list[idx-1]:
    # count: 현재 무게의 공이 몇 개?
    # (n - idx): 현재 무게 초과하는 공이 몇 개?
    answer += count * (n - idx)
    count = 1
  else:
    count += 1

print(answer)
```

* 모범답안은 `1 <= M <= 10`인 걸 고려하여, 공별 개수 array를 만듦
* 다만 `(현재 무게의 공 개수) * (현재 무게 초과하는 공 개수)` 곱해서 총 개수 세는 것은 동일
* 해쉬테이블 사용이 다소 효율적이긴 하겠다

```python
# 입력
n, m = map(int, input().split()) # 볼링공 개수, 최대무게
ball_list = list(map(int, input().split()))
ball_count = [0] * (m + 1) # 무게별 공 개수
for b in ball_list:
  ball_count[b] += 1

answer = 0

for idx in range(1, m + 1):
  n -= ball_count[idx] # 현재 무게 초과하는 공 개수에서 빼주기
  answer += ball_count[idx] * n # 그리고 곱해서 더하기

print(answer)
```
