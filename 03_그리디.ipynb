# [예제] 거스름돈



```python
# 나의 답안
money = int(input())
m_list = [500, 100, 50, 10]
result = 0

for m in m_list:
  amount = money // m
  result += amount
  money -= amount * m

print(result)
```

* **가장 큰 화폐 단위**부터 돈을 거슬러 준다.

* 시간 복잡도: 화폐 종류가 $K$개일 때, $O(K)$

* 더 큰 단위의 동전이 항상 작은 단위의 배수이므로, 작은 단위의 동전들을 종합해 다른 해가 나올 수 없다 -> 그리디 알고리즘 사용 가능

```python
# 모범답안
money = int(input())
result = 0

for m in [500, 100, 50, 10]:
  result += money // m
  money %= m

print(result)
```

# [실전] 큰 수의 법칙

```python
# 나의 답안
n, m, k = map(int, input().split())
numbers = list(map(int, input().split()))
numbers.sort(reverse = True)

times2 = m // (k + 1)
times1 = m - times2
answer = numbers[0] * times1 + numbers[1] * times2

print(answer)
```

* 입력값 중 가장 큰 수와 두번째로 큰 수만 저장한다.
* 가장 큰 수를 K번 더하고, 두번째로 큰 수를 1번 더하는 연산을 반복한다.

```python
# 모범답안 1
n, m, k = map(int, input().split())
numbers = list(map(int, input().split()))
numbers.sort()
result = 0

while True:
  for _ in range(k):
    if m == 0:
      break
    result += numbers[n - 1]
    m -= 1
  if m == 0:
    break
  result += numbers[n - 2]
  m -= 1

print(result)
```

* 반복되는 수열을 반복한다.
* (제일 큰 숫자) `K`번 -> (2번째로 큰 숫자) 1번, 즉 각 수열의 길이는 `K + 1`
* 즉 수열은 `M // (K + 1)`번 반복. 이후 나머지만큼 가장 큰 수가 추가적으로 더해짐.


* 가장 큰 수가 더해지는 횟수
`(M // (K + 1)) * K + M % (K + 1)`



```python
# 모범답안 2
n, m, k = map(int, input().split())
numbers = list(map(int, input().split()))
numbers.sort()

count = (m // (k + 1)) * k # 가장 큰 수가 더해지는 횟수
count += m % (k + 1)

result = 0
result += (count) * numbers[n - 1]
result += (m - count) * numbers[n - 2]

print(result)
```

# [실전] 숫자 카드 게임

```python
# 나의 답안
n, m = map(int, input().split())
cards = []
row_mins = []

for _ in range(n):
  cards.append(list(map(int, input().split())))

answer = 0

for idx in range(n):
  row_mins.append(min(cards[idx]))

answer = max(row_mins)

print(answer)
```

# [실전] 1이 될 때까지

```python
# 나의 풀이
n, k = map(int, input().split())
answer = 0

while n > 1:
  if n % k == 0:
    n //= k
  else:
    n -= 1
  answer += 1

print(answer)
```

* 최대한 많이 나누기 - 2 이상으로 나누는 것이, 1을 빼는 것보다 숫자를 많이 줄일 수 있다
* `K`가 2 이상이기만 하면 `K`로 나누는 것이 항상 더 효율적이다

즉, 아래와 같이 수행
1. `N`이 `K`의 배수가 될 때까지 1 빼기
2. `N`을 `K`로 나누기

```python
# 모범답안
n, k = map(int, input().split())
answer = 0

while True:
  # (N == K로 나누어떨어지는 수)가 될 때까지 1씩 빼기
  target = (n // k) * k
  answer += (n - target)
  n = target

  # N이 K보다 작을 때 반복문 탈출
  if n < k:
    break

  answer += 1
  n //= k

# 마지막으로 남은 수에 대해 1씩 빼기
answer += (n - 1)
print(answer)
```
