# Q41 여행 계획

```python
n, m = map(int, input().split())
parent = [i for i in range(n + 1)]

def find(parent, x):
    if parent[x] != x:
        parent[x] = find(parent, parent[x])
    return parent[x]

def union(parent, a, b):
    a = find(parent, a)
    b = find(parent, b)
    if a < b:
        parent[b] = a
    else:
        parent[a] = b

for x in range(1, n + 1):
    line = list(map(int, input().split()))
    for y in range(1, n + 1):
        if line[y - 1] == 1:
            union(parent, x, y)

plan = list(set(map(int, input().split())))

result = set([find(parent, p) for p in plan])
print(parent)

if len(result) == 1:
    print("YES")
else:
    print("NO")

```

# Q42 탑승구

```python
def find(parent, x):
    if parent[x] != x:
        parent[x] = find(parent, parent[x])
    return parent[x]

def union(parent, a, b):
    a = find(parent, a)
    b = find(parent, b)
    if a < b:
        parent[b] = a
    else:
        parent[a] = b

g = int(input())
p = int(input())
parent = [i for i in range(g + 1)]
out = []

for _ in range(p):
    out.append(int(input()))

answer = 0

for o in out:
    # 현재 비행기의 탑승구의 루트 확인
    curr = find(parent, o)
    # 현재 루트가 0인 경우 종료
    if curr == 0:
        break
    # 그렇지 않으면 왼쪽 집합과 합치기
    union(parent, curr, curr - 1)
    answer += 1

print(answer)
```

* 가능한 큰 번호의 탑승구를 찾기 위해선 집합의 루트를 확인해야 한다. (`find`) 도킹을 수행한다.
* 도킹을 수행하는 탑승구와, 앞 탑승구를 한 집합으로 합친다(`union`)
* 단, 집합의 루트가 0인 경우 도킹이 불가능한 것으로 판단한다.

# Q43 어두운 길

```python
n, m = map(int, input().split())
edges = []
parent = [i for i in range(n)]

for _ in range(m):
    x, y, cost = map(int, input().split())
    edges.append((cost, x, y))

edges.sort()

def find(parent, x):
    if parent[x] != x:
        parent[x] = find(parent, parent[x])
    return parent[x]

def union(parent, a, b):
    a = find(parent, a)
    b = find(parent, b)

    if a < b:
        parent[b] = a
    else:
        parent[a] = b

answer = 0

for cost, x, y in edges:
    if find(parent, x) != find(parent, y):
        union(parent, x, y)
    else:
        answer += cost

print(answer)
```

# Q45 최종 순위

```python
from collections import deque
# import sys
# input = lambda: sys.stdin.readline().rstrip()

def topsort(graph, degree):
    queue = deque([])
    result = []

    for i in range(1, n + 1):
        if degree[i] == 0:
            queue.append(i)

    while queue:
        if len(list(queue)) >= 2:
            return "?"
        curr = queue.popleft()
        result.append(curr)
        for next in graph[curr]:
            degree[next] -= 1
            if degree[next] == 0:
                queue.append(next)

    if len(result) == n:
        return " ".join([str(i) for i in result])
    else:
        return "IMPOSSIBLE"

t = int(input())

for _ in range(t):
    n = int(input())
    lastyear = [0] + list(map(int, input().split()))
    m = int(input())
    changed = []
    for _ in range(m):
        changed.append(list(map(int, input().split())))

    graph = [[] for _ in range(n + 1)]
    degree = [0] * (n + 1)

    # 그래프 방향은 높 -> 낮.

    # 높은 등수
    for i in range(1, n + 1):
        # 낮은 등수
        for j in range(i + 1, n + 1):
            high = lastyear[i]
            low = lastyear[j]

            if [high, low] in changed or [low, high] in changed:
                graph[low].append(high)
                degree[high] += 1
            else:
                graph[high].append(low)
                degree[low] += 1
    print(topsort(graph, degree))
```
