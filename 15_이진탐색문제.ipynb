# Q27 정렬된 배열에서 특정 수의 개수 구하기

```python
import bisect

n, x = map(int, input().split())
array = list(map(int, input().split()))
answer = bisect.bisect_right(array, x) - bisect.bisect_left(array, x)

if answer < 1:
    print(-1)
else:
    print(answer)
```

* $O(log N)$으로 동작하는 알고리즘 -> 모든 원소가 정렬되어 있으니, 이진 탐색 사용 가능
* 일일이 이진탐색 함수를 쓰는 방법도 있지만, 여기선 `bisect` 쓰는게 이득

# Q28 고정점 찾기

```python
n = int(input())
array = list(map(int, input().split()))
left, right = 0, n - 1

while left <= right:
    mid = (left + right) // 2
    if mid == array[mid]:
        print(mid)
        break
    elif mid < array[mid]:
        right = mid - 1
    else:
        left = mid + 1
else:
    print(-1)
```

* 역시 $O(log N)$에다 리스트가 정렬되어 있음 -> 이진 탐색 수행
* 중간점이 가리키는 위치의 값보다 중간점이 작은 경우 -> 왼쪽 부분 탐색
* 중간점이 가리키는 위치의 값보다 중간점이 큰 경우 -> 오른쪽 부분 탐색

# Q29 공유기 찾기

```python
n, c = map(int, input().split())
houses = []
for _ in range(n):
    houses.append(int(input()))
houses.sort()
answer = 0

def check_dist(array, start, end, rem):
    global answer
    if len(array[start:end]) == 1:
        return 1e9
    elif len(array[start:end]) == 2:
        return array[1] - array[0]
    else:
        for cut in range(start + 1, end):
            left_dist = check_dist(array, start, cut)
            right_dist = check_dist(array, cut, end)
            inter_dist = array[cut] - array[cut - 1]
            print(start, cut, end, left_dist, right_dist, inter_dist)
            max(answer, min(left_dist, right_dist, inter_dist))
        return answer

print(check_dist(houses, 0, n))
```
