# 자료구조 기초

## 스택
* First In Last Out / Last In First Out
* 삭제 시, 마지막으로 추가한 데이터가 삭제된다

```python
# 별도의 라이브러리를 사용하지 않아도 됨.

stack = []
stack.append(5) # [5]
stack.append(2) # [5, 2]
stack.append(3) # [5, 2, 3]
stack.append(7) # [5, 2, 3, 7]
stack.pop() # [5, 2, 3]
stack.append(1) # [5, 2, 3, 1]
stack.append(4) # [5, 2, 3, 1, 4]
stack.pop() # [5, 2, 3, 1]

print(stack) # 최하단 원소부터 출력
print(stack[::-1]) # 최상단 원소부터 출력
```

## 큐

* First In First Out
* 삭제 시, 제일 먼저 추가한 데이터가 삭제된다

```python
# 삭제 시 deque 라이브러리의 popleft() 메서드를 사용해야 함.

from collections import deque
queue = deque()

queue.append(5) # [5]
queue.append(2) # [5, 2]
queue.append(3) # [5, 2, 3]
queue.append(7) # [5, 2, 3, 7]
queue.popleft() # [2, 3, 7]
queue.append(1) # [2, 3, 7, 1]
queue.append(4) # [2, 3, 7, 1, 4]
queue.popleft() # [3, 7, 1, 4]

print(queue) # 먼저 들어온 순서대로 출력
queue.reverse() # deque는 slicing이 안 됨에 유의
print(queue) # 나중에 들어온 원소부터 출력
```

## 재귀 함수
* 자기 자신을 다시 호출하는 함수
* **종료 조건**: 재귀 함수가 언제 끝날지 명시해야 함 - 무한 호출을 막기 위함

```python
# 재귀의 최대 깊이를 초과할 때까지 계속 print

def recursive_function(i):
    if i == 100: # 종료조건
        return
    print(f"{i}번째 재귀 함수에서 {i+1}번째 재귀 함수를 호출합니다.")
    recursive_function(i + 1)
    print(f"{i}번째 재귀 함수를 종료합니다.")

recursive_function(1)
```

* 재귀함수의 수행은 **스택**을 이용
  * 가장 마지막에 호출한 함수가 먼저 수행을 끝내야, 그 앞의 함수 호출이 종료됨
  * 스택을 활용해야 하는 알고리즘은 재귀 함수를 이용해 간단히 구현 가능 (DFS)

```python
# 팩토리얼

# 반복적 구현
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# 재귀적 구현
def factorial_recursive(n):
    if n <= 1: # 종료조건
        return 1
    # n! = n * (n - 1)!
    return n * factorial_recursive(n - 1)

print(f"반복적으로 구현: {factorial_iterative(5)}")
print(f"재귀적으로 구현: {factorial_recursive(5)}")
```

## 그래프
* 그래프의 기본 구조
  * 노드(node)와 간선(edge)으로 구분
  * 그래프 탐색: 하나의 노드를 시작으로 다수의 노드를 반복하는 것
  * 두 노드가 간선으로 연결되어 있다 -> 두 노드는 인접하다
* 그래프의 표현 방식
  * 인접 행렬(Adjacency Matrix): 2차원 배열로 연결관계 표현
  * 인접 리스트(Adjacency List): 리스트로 연결관계 표현

```python
# 인접 행렬
import math
INF = math.inf

# 2차원 리스트를 이용해 인접 행렬 표현
# 0: 동일 노드. INF: 연결되어 있지 않음.
graph = [
    [0, 7, 5],
    [7, 0, INF],
    [5, INF, 0]
]

print(graph)
```

```python
# 인접 리스트

# 행이 3개인 2차원 리스트로 인접 리스트 표현
graph = [[] for _ in range(3)]

# 노드 0에 연결된 노드 정보: (노드, 거리)
graph[0].append((1, 7))
graph[0].append((2, 5))

# 노드 1에 연결된 노드 정보: (노드, 거리)
graph[1].append((0, 7))

# 노드 2에 연결된 노드 정보: (노드, 거리)
graph[2].append((0, 5))

print(graph)
```

* 두 방식의 비교
  * 인접 행렬
    * 모든 관계를 저장하므로 노드 개수가 많을수록 메모리가 불필요하게 낭비된다.
    * 특정한 두 노드의 연결 여부를 빠르게 확인할 수 있다. e.g., `graph[1][7]` 확인
  * 인접 리스트
    * 연결된 정보만을 저장하기 때문에 메모리를 효율적으로 사용한다.
    * 연결된 데이터를 하나씩 확인해야 하기 때문에 인접 행렬 방식에 비해 특정한 두 노드가 연결되어 있는지에 대한 정보를 얻는 속도가 느리다. e.g., `graph[1]`을 차례대로 확인

## DFS (Depth First Search)
* 깊이 우선 탐색, 그래프에서 깊은 부분을 우선적으로 탐색
* 스택 자료구조 사용, 재귀 함수 사용 가능
* DFS의 절차
  1. 탐색 시작 노드를 스택에 삽입하고 방문 처리를 한다.
  2. 스택의 최상단 노드에 방문하지 않은 인접 노드가 있다면, **인접 노드 중 1개**를 스택에 넣고 방문 처리를 한다. 방문하지 않은 인접 노드가 없으면, 스택에서 최상단 노드를 꺼낸다. (보통은 관행적으로 번호가 제일 낮은 걸 넣음)
  3. (2)를 더 이상 수행할 수 없을 때까지 반복한다.
* 데이터의 개수가 $N$개일 때 $O(N)$

```python
# 재귀함수를 이용한 DFS
def dfs(graph, v, visited):
    # 현재 노드를 방문
    visited[v] = True
    print(v, end=" ")

    # 현재 노드와 연결된 다른 노드를 재귀적으로 방문
    for i in graph[v]:
        # 스택의 최상단 노드에 방문하지 않은 인접 노드가 있다면,
        if not visited[i]:
            # 스택에 넣고(재귀함수를 실행하고) 방문 처리를 한다.
            dfs(graph, i, visited)
    # 방문하지 않은 인접 노드가 없으면, 스택에서 최상단 노드를 꺼낸다(재귀함수가 종료된다)

# 그래프를 인접 리스트로 표현
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

# 각 노드가 방문된 정보를 1차원 리스트로 표현
visited = [False] * 9

# 정의된 DFS 함수 호출
# 탐색 시작 노드를 스택에 삽입하고(함수를 호출하고) 방문 처리를 하기
dfs(graph, 1, visited)
```

## BFS (Breadth First Search)
* 너비 우선 탐색, 가까운 노드부터 탐색하는 알고리즘
* 큐 자료구조 사용 (`deque` 라이브러리)
* BFS의 절차
  1. 탐색 시작 노드를 큐에 삽입하고 방문 처리를 한다.
  2. 큐에서 노드를 꺼내 해당 노드의 인접 노드 중에서 방문하지 않은 노드를 **모두** 큐에 삽입하고 방문 처리를 한다.
  3. (2)를 더 이상 수행할 수 없을 때까지 반복한다.
* $O(N)$, 일반적으로 DFS보다 수행 시간 좋은 편

```python
# deque를 이용한 BFS
from collections import deque

# BFS 메서드 정의
def bfs(graph, start, visited):
    # 탐색 시작 노드를 큐에 삽입하고 방문 처리
    queue = deque([start])
    visited[start] = True

    # 큐에서 노드를 꺼낼 수 있는 동안 반복
    while queue:
        # 큐에서 노드를 꺼내기
        v = queue.popleft()
        print(v, end=" ")

        # 인접 노드 중에서 방문하지 않은 노드를 모두 큐에 삽입하고, 방문 처리하기
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

# 그래프를 인접 리스트로 표현
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

# 각 노드가 방문된 정보를 1차원 리스트로 표현
visited = [False] * 9

# 정의된 DFS 함수 호출
bfs(graph, 1, visited)
```

# [실전] 음료수 얼려 먹기

```python
# 1차 시도
from collections import deque

def bfs(graph, start, unvisited):
    queue = deque([start])
    unvisited.remove(start)

    while queue:
        node = queue.popleft()
        for next in graph[node]:
            if next in unvisited:
                queue.append(next)
                unvisited.remove(next)

    return unvisited

def node_no(x, y, m):
    return (x - 1) * m + y

n, m = map(int, input().split())
grid = [[1] * (m + 2) for _ in range(n + 2)]
graph = dict()
unvisited = []

for x in range(1, n + 1):
    grid[x] = [1] + list(map(int, input())) + [1]

for x in range(1, n + 1):
    for y in range(1, m + 1):
        if grid[x][y] == 0:
            curr = node_no(x, y, m)
            graph[curr] = []
            unvisited.append(curr)
            if grid[x - 1][y] == 0: # 상단
                graph[curr].append(node_no(x - 1, y, m))
            if grid[x + 1][y] == 0: # 하단
                graph[curr].append(node_no(x + 1, y, m))
            if grid[x][y - 1] == 0: # 좌측
                graph[curr].append(node_no(x, y - 1, m))
            if grid[x][y + 1] == 0: # 우측
                graph[curr].append(node_no(x, y + 1, m))

answer = 0
while unvisited:
    start = unvisited[0]
    unvisited = bfs(graph, start, unvisited)
    answer += 1

print(answer)
```

```python
# 2차 시도 -> 그리드 서치하면서 bfs 시도?
from collections import deque

n, m = map(int, input().split())
grid = [[1] * (m + 2) for _ in range(n + 2)]

for x in range(1, n + 1):
    grid[x] = [1] + list(map(int, input())) + [1]

answer = 0

for x in range(1, n + 1):
    for y in range(1, m + 1):
        if grid[x][y] == 0:
            queue = deque([(x, y)])
            grid[x][y] = 1

            while queue:
                cx, cy = queue.popleft()
                for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                    nx, ny = cx + dx, cy + dy
                    if grid[nx][ny] == 0:
                        queue.append((nx, ny))
                        grid[nx][ny] = 1

            answer += 1

print(answer)
```

```python
# 3차 시도 -> dfs를 시도한다면?

def dfs(x, y):
    if grid[x][y] == 0:
        grid[x][y] = 1

        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            dfs(x + dx, y + dy)
        return True
    return False

n, m = map(int, input().split())
grid = [[1] * (m + 2) for _ in range(n + 2)]

for x in range(1, n + 1):
    grid[x] = [1] + list(map(int, input())) + [1]

answer = 0

for x in range(1, n + 1):
    for y in range(1, m + 1):
        if dfs(grid, x, y):
            answer += 1

print(answer)
```

# [실전] 미로 탈출

```python
from collections import deque

def bfs(x, y, d):
    queue = deque([(x, y, 1)]) # queue에 삽입
    grid[x][y] = 0 # 방문 처리

    while queue:
        cx, cy, cd = queue.popleft() # 인출
        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nx, ny = cx + dx, cy + dy

            # 목적지에 도달
            if nx == n and ny == m:
                return cd + 1

            # 근접한 곳 중 갈 수 있는 곳을 큐에 추가, 방문 처리
            elif grid[nx][ny] == 1:
                queue.append((nx, ny, cd + 1))
                grid[nx][ny] = 0

n, m = map(int, input().split())
grid = [[0] * (m + 2) for _ in range(n + 2)]

for rdx in range(1, n + 1):
    grid[rdx] = [0] + list(map(int, input())) + [0]

answer = bfs(1, 1, 1)
print(answer)
```

* 어떻게든 depth를 계속 기록해야 하는데, 나는 튜플을 3차원으로 만들었고, 모범답안은 아예 grid에 depth를 계속 표시해 나감.

```python
# 모범답안

from collections import deque

def bfs(x, y):
    queue = deque([(x, y)]) # queue에 삽입

    while queue:
        cx, cy = queue.popleft() # 인출
        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nx, ny = cx + dx, cy + dy

            # 근접한 곳 중 갈 수 있는 곳을 큐에 추가, 깊이 기록 처리
            if grid[nx][ny] == 1:
                queue.append((nx, ny))
                grid[nx][ny] = grid[cx][cy] + 1 # 깊이 더하기

    print(grid)
    return grid[n][m]

n, m = map(int, input().split())
grid = [[0] * (m + 2) for _ in range(n + 2)]

for rdx in range(1, n + 1):
    grid[rdx] = [0] + list(map(int, input())) + [0]

answer = bfs(1, 1)
print(answer)
# 첫칸은 1이 아니라 3으로 찍히지만, 이 문제 특성상 상관없다!
```
