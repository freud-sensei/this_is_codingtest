# Q7 럭키 스트레이트

https://www.acmicpc.net/problem/18406

```python
n = input()

# 좌우로 나누기
center = len(n) // 2
left = n[:center]
right = n[center:]

# 자리수 합 더하기
left_sum = sum(int(c) for c in left)
right_sum = sum(int(c) for c in right)

# 판정
if left_sum == right_sum:
  print("LUCKY")
else:
  print("READY")
```

* 그냥 문제에서 요구하는 대로 하면 된다.

```python
# 모범답안
n = input()
summary = 0

# 왼쪽 더하기
for i in range(len(n) // 2):
  summary += int(n[i])

# 오른쪽 빼기
for i in range(len(n) // 2, len(n)):
  summary -= int(n[i])

# 판정
if summary == 0:
  print("LUCKY")
else:
  print("READY")
```

# Q9 문자열 압축

https://programmers.co.kr/learn/courses/30/lessons/60057

```python
# 나의 답안

def solution(s):
    # 압축을 하지 않았을 때
    answer = len(s)

    # 문자열을 몇 개 단위로 자르기?
    for cut_length in range(1, len(s) // 2 + 1):
        result = []
        count = 1 # 지금 자른 게 몇 번째 반복?

        # 이전 cut과 비교
        for idx in range(0, len(s) - cut_length, cut_length):
            prev_cut = s[idx: idx + cut_length]
            cut = s[idx + cut_length: idx + cut_length * 2]

            # 이전 cut과 같을 때
            if prev_cut == cut:
                count += 1

            # 이전 cut과 다를 때
            else:
                if count == 1:
                    result.append(prev_cut)
                else:
                    result.append(f"{count}{prev_cut}")
                    count = 1

        # 마지막 cut 추가
        if count == 1:
            result.append(cut)
        else:
            result.append(f"{count}{cut}")

        result = "".join(result)

        if len(result) < answer:
            answer = len(result)

    return answer
```

* 반복문에서 이전 반복의 변수가 그대로 넘어갈 수 있음을 염두에 두자.
* 문자열의 길이가 1,000 이하이므로 완전 탐색을 수행해도 무방하다.
* `1`부터 `N/2`까지만의 모든 수를 단위로 하면 된다. 넘치는 범위는 고려할 필요 없음.

```python
# 나의 답안

def solution(s):
    # 압축을 하지 않았을 때
    answer = len(s)

    # 문자열을 몇 개 단위로 자르기?
    for cut_length in range(1, len(s) // 2 + 1):
        result = ""
        count = 1 # 지금 자른 게 몇 번째 반복?
        prev = s[0: cut_length]

        # 이전 cut과 비교
        for idx in range(cut_length, len(s), cut_length):
            cut = s[idx: idx + cut_length]

            # 이전 cut과 같을 때
            if prev == cut:
                count += 1

            # 이전 cut과 다를 때
            else:
                if count > 1:
                    result += f"{count}{prev}"
                else:
                    result += prev
                count = 1

        # 마지막 cut 추가
        if count > 1:
            result += f"{count}{prev}"
        else:
            result += prev

        answer = min(answer, len(result))

    return answer
```

# Q10 자물쇠와 열쇠
https://programmers.co.kr/learn/courses/30/lessons/60059

```python
# 열쇠를 90도 회전
def rotate(m, key_1):
    rotated_key = []
    for row, col in key_1:
        rotated_key.append((col, m - 1 - row))
    return rotated_key

def solution(key, lock):
    m = len(key)
    n = len(lock)
    key_1 = []
    lock_0 = set()

    # key의 돌기 부분 원소
    for r in range(m):
        for c in range(m):
            if key[r][c] == 1:
                key_1.append((r, c))

    # lock의 홈 부분 원소
    for r in range(n):
        for c in range(n):
            if lock[r][c] == 0:
                # 자물쇠를 확장하기 위해 n 추가
                lock_0.add((r + n, c + n))

    # 열쇠의 회전
    for _ in range(4):

        # 열쇠의 위치
        for key_r in range(2 * n):
            for key_c in range(2 * n):
                key_input = set()

                for fit in key_1:
                    input_r = fit[0] + key_r
                    input_c = fit[1] + key_c
                    if n <= input_r < 2 * n and n <= input_c < 2 * n:
                        key_input.add((input_r, input_c))

                # 맞는 경우가 있는 경우
                if lock_0 == key_input:
                    return True

        # 맞는 경우가 없는 경우, 90도 회전
        key_1 = rotate(m, key_1)

    # 모든 회전에서 끼워넣을 수 없는 홈이 존재
    return False
```

* 슬슬 구현 유형이 화나기 시작한다...
* 좌물쇠, 열쇠의 크기가 20x20보다 작으므로, 완전 탐색을 사용해도 좋다.
* 모범답안에선 좌물쇠를 크기가 3배인 새로운 리스트로 만들었고, 왼쪽 위부터 한 칸씩 이동해 모든 홈을 채울 수 있는지 확인함. 내 풀이처럼 어정쩡한 숫자 들어가는 것보단, 3으로 맞추는 게 편리한 듯.
* 좌물쇠 리스트 + 열쇠 리스트 한 뒤, 모든 값이 정확히 1이면 다 채운 것.
* 90도 돌리기를 다른 함수로 구현한 건 아주 잘 하셨고요

```python
# 열쇠를 90도 회전
def rotate(key):
    m = len(key)
    result = [[0 * m for _ in range(m)]]

    for idx in range(m):
      for jdx in range(m):
        result[jdx][m - idx - 1] = key[idx][jdx]

    return result

# 자물쇠의 중간 부분이 모두 1인지 확인
def check(new_lock):
  n = len(new_lock) // 3
  for idx in range(n, n * 2):
    for jdx in range(n, n * 2):
      if new_lock[idx][jdx] != 1:
        return False
  return True

def solution(key, lock):
    m = len(key)
    n = len(lock)

    # 자물쇠의 크기를 3배로 변환
    new_lock = [[0] * (n * 3) for _ in range(n * 3)]
    for i in range(n):
      for j in range(n):
        new_lock[i + n][j + n] = lock[i][j]

    # 열쇠의 회전
    for _ in range(4):
      key = rotate(key)
      for x in range(n ** 2):
        for y in range(n ** 2):
          for i in range(m):
            for j in range(m):
              new_lock[x + i][y + j] += key[i][j]

          if check(new_lock):
            return True

          for i in range(m):
            for j in range(m):
              new_lock[x + i][y + j] -= key[i][j]



    # 모든 회전에서 끼워넣을 수 없는 홈이 존재
    return False
```

# Q11 뱀

https://www.acmicpc.net/problem/3190

```python
# 입력
n = int(input()) # 보드크기
k = int(input()) # 사과 수
apples = []
for _ in range(k):
    apples.append(list(map(int, input().split())))
l = int(input())
dir_change = dict()
for _ in range(l):
    x, c = input().split()
    dir_change[int(x)] = c

# 게임 구현
snake_length = 1
time = 1
directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
snake_loc = [1, 1]
facing = 0
body = [[1, 1]]

# 게임 진행
while True:
    # 뱀이 이동
    snake_loc[0] += directions[facing][0]
    snake_loc[1] += directions[facing][1]

    # 벽에 부딪힐 때
    if snake_loc[0] < 1 or snake_loc[0] > n or snake_loc[1] < 1 or snake_loc[1] > n:
        break

    # 자신의 몸에 부딪힐 때
    if snake_loc in body:
        break

    # 사과가 있을 때
    if snake_loc in apples:
        apples.remove(snake_loc)
        snake_length += 1

    # body에 이동정보 추가
    body.append(snake_loc.copy())
    if len(body) > snake_length:
        body.pop(0)

    # 해당 시간마다 방향을 전환
    if time in dir_change:
        if dir_change[time] == "L":
            facing = facing - 1 if facing != 0 else 3
        else:
            facing = facing + 1 if facing != 3 else 0

    # 모든 과정이 끝나면 1초 추가
    time += 1

print(time)
```

* **시뮬레이션** 유형: 시키는 대로 하세요!
* 매 시점마다 뱀이 존재하는 위치를 항상 기록해야 합니다. 난 그냥 리스트에다 기록했고, 모범답안은 2차원 리스트를 만들어서 진행함. FIFO 형태로 pop 하면서 계속 관리... 큐 기억하죠?
* index 변수를 따로 두어 언제 회전할지 체크.

```python
# 모범답안2

# 입력
n = int(input()) # 보드크기
k = int(input()) # 사과 수

board = [[0] * (n + 1) for _ in range(n + 1)]
for _ in range(k):
    row, col = map(int, input().split())
    board[row][col] = 1

l = int(input())
dir_change = []
for _ in range(l):
    x, c = input().split()
    dir_change.append((int(x), c))

# 방향 설정
dx = [0, 1, 0, -1]
dy = [1, 0, -1, 0]

# 돌기
def turn(direction, c):
    if c == "L":
        direction = (direction - 1) % 4
    else:
        direction = (direction + 1) % 4
    return direction

# 게임 구현
def simulate():
    snake_x = 1
    snake_y = 1
    board[snake_x][snake_y] = 2 # 뱀이 차지하는 위치정보
    time = 0 # 시작 후 지난 시간
    index = 0 # 언제 회전?
    facing = 0
    body = [(snake_x, snake_y)] # 뱀이 차지하는 위치정보, FIFO

    # 게임 진행
    while True:
        # 뱀이 이동
        nx = snake_x + dx[facing]
        ny = snake_y + dy[facing]

        # 벽을 안 친 경우 and 내 몸을 안 친 경우
        if 1 <= nx <= n and 1 <= ny <= n and board[nx][ny] != 2:
            # 사과 X
            if board[nx][ny] == 0:
                board[nx][ny] = 2
                body.append((nx, ny))
                px, py = body.pop(0)
                board[px][py] = 0
            # 사과 O
            if board[nx][ny] == 1:
                board[nx][ny] = 2
                body.append((nx, ny))
        else:
            time += 1
            break
        # 위치 이동
        snake_x, snake_y = nx, ny
        time += 1

        # 회전할 시간이 되었을 때
        if index < l and time == dir_change[index][0]:
            facing = turn(facing, dir_change[index][1])
            index += 1

    return time

print(simulate())
```
