# Q23 국영수

```python
n = int(input())
students = []

for _ in range(n):
    name, s1, s2, s3 = input().split()
    students.append((name, int(s1), int(s2), int(s3)))

students.sort(key=lambda x: (-x[1], x[2], -x[3], x[0]))

for s in students:
    print(s[0])
```

* `key`에 `lambda` 함수 넣을 때 역순으로 숫자를 정렬하고 싶으면 음수로 처리

# Q24 안테나

```python
n = int(input())
houses = list(map(int, input().split()))
houses.sort()

if n % 2 == 1:
    print(houses[n // 2])
else:
    left = houses[n // 2 - 1]
    right = houses[n // 2]
    left_sum = sum(abs(left - h) for h in houses)
    right_sum = sum(abs(right - h) for h in houses)

    if left_sum <= right_sum:
        print(left)
    else:
        print(right)
```

* 이게 진짜 정답이라고? 정렬의 탈을 쓴 수학 문제 같다.
* 놀랍게도 중앙값에 설치하면 총합이 최소가 된다.
* `n`이 짝수인 경우 두 값 중 더 작은 값에 설치하면 된다.
* 놀랍게도 항상 성립한다.

```python
n = int(input())
houses = list(map(int, input().split()))
houses.sort()

if n % 2 == 1:
    print(houses[n // 2])
else:
    print(houses[n // 2 - 1])
```

# Q25 실패율

```python
def solution(N, stages):
    fail_users = [0] * (N + 1)
    fail_rate = [0] * (N + 1)
    users = len(stages)

    # 해당 스테이지 도전자 수
    for s in stages:
        if s <= N:
            fail_users[s] += 1

    # 실패율 계산
    for s in range(1, N + 1):
        if users <= 0:
            fail_rate[s] = 0
        else:
            fail_rate[s] = fail_users[s] / users
            users -= fail_users[s]

    # 정답 출력
    answer = [s for s in range(1, N + 1)]
    answer.sort(key=lambda s:(-fail_rate[s]))

    return answer

solution(5, [2, 1, 2, 6, 2, 4, 3, 3])
```

* 계수 정렬 같으면서도 뭔가 아닌 것 같다...? 일단 실패율을 $O(N + K)$로 계산하는 것이 중요한 듯.
* 구현에 더 가까운 문제. 이때 해당 스테이지에 도달한 플레이어가 없으면 실패율이 0이 됨에 유의할 것!

```python
# 모범답안
def solution(N, stages):
    answer = []
    users = len(stages)

    # 실패율 계산
    for s in range(1, N + 1):
        fail_users = stages.count(s)
        if fail_users == 0:
            fail = 0
        else:
            fail = fail_users / users
        answer.append((s, fail))
        users -= fail_users

    # 정답 출력
    answer.sort(key=lambda x: x[1], reverse=True)
    answer = [x[0] for x in answer]
    return answer

solution(5, [2, 1, 2, 6, 2, 4, 3, 3])
```

# Q26 카드 정렬하기

```python
import heapq

n = int(input())
cards = []

for _ in range(n):
    heapq.heappush(cards, int(input()))
answer = 0

while len(cards) > 1:
    c1, c2 = heapq.heappop(cards), heapq.heappop(cards)
    answer += c1 + c2
    heapq.heappush(cards, c1 + c2)

print(answer)
```

* 우선순위 큐... 이젠 확실히 알아야 한다.
* 그리디 문제에 가깝다. 즉 항상 가장 작은 크기의 두 카드 묶음을 합쳤을 때 최적의 해를 보장한다.
