# 구현
* **완전 탐색**: 모든 경우의 수를 다 계산한다.
* **시뮬레이션**: 문제에서 제시한 알고리즘을 한 단계씩 차례대로 직접 수행한다.

# [예제] 상하좌우

```python
# 나의 답안
n = int(input())
move_list = input().split()
x, y = 1, 1

for m in move_list:
  if m == "L" and y > 1:
    y -= 1
  elif m == "R" and y < n:
    y += 1
  elif m == "U" and x > 1:
    x -= 1
  elif m == "D" and x < n:
    x += 1

print(x, y)
```

* 시뮬레이션 유형: 일련의 명령에 따라 개체를 차례대로 이동시킨다.
* dx, dy를 리스트에 넣어서 정리하면 깔끔!

```python
# 모범 답안
n = int(input())
move_list = input().split()
x, y = 1, 1

dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
move_types = ["L", "R", "U", "D"]

for m in move_list:
  for i in range(4):
    if m == move_types[i]:
      nx = x + dx[i]
      ny = y + dy[i]
  if nx < 1 or ny < 1 or nx > n or ny > n:
    continue
  x, y = nx, ny

print(x, y)
```

# [예제] 시각

```python
# 나의 답안
n = int(input())
answer = 0

for hour in range(n + 1):
  for minute in range(60):
    for second in range(60):
      if "3" in str(hour) or "3" in str(minute) or "3" in str(second):
        answer += 1

print(answer)
```

* 완전 탐색 유형: 하루는 86,400초로 경우의 수가 100,000개도 되지 않으므로 쉽게 시간 제한 내에 문제 해결 가능.
* 보통 전체 데이터 수가 100만개 이하일 때 완전 탐색을 사용하는 것이 적절.

```python
# 모범 답안
n = int(input())
answer = 0

for hour in range(n + 1):
  for minute in range(60):
    for second in range(60):
      if "3" in str(hour) + str(minute) + str(second):
        answer += 1

print(answer)
```

# [예제] 왕실의 나이트

```python
# 좌표 입력받기
col_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4,
            'e': 5, 'f': 6, 'g': 7, 'h': 8}
start = input()
r = int(start[1])
c = col_dict[start[0]]
answer = 0

# 움직일 수 있는 경로 - (row, column)
moves = [(-1, -2), (-1, 2), (1, -2), (1, 2), (-2, -1), (-2, 1), (2, -1), (2, 1)]

# 밖으로 안 나가나 체크
for m in moves:
  new_r = r + m[0]
  new_c = c + m[1]
  if 1 <= new_r and new_r <= 8 and 1 <= new_c and new_c <= 8:
    answer += 1

print(answer)
```

* <상하좌우>와 비슷한 시뮬레이션 유행, 이동 경로를 하나씩 확인한 뒤 이에 맞춰 이동
* 알파벳 순서와 같은 경우 `ord('a')` 함수를 이용할 수 있다.

```python
# 모범답안 수정

# 좌표 입력받기
start = input()
r = int(start[1])
c = ord(start[0]) - ord('a') + 1
answer = 0

# 움직일 수 있는 경로 - (row, column)
moves = [(-1, -2), (-1, 2), (1, -2), (1, 2), (-2, -1), (-2, 1), (2, -1), (2, 1)]

# 밖으로 안 나가나 체크
for m in moves:
  new_r = r + m[0]
  new_c = c + m[1]
  if 1 <= new_r and new_r <= 8 and 1 <= new_c and new_c <= 8:
    answer += 1

print(answer)
```

# [실전] 게임 개발

```python
# 입력 및 맵 생성
# d: 0 북쪽, 1 동쪽, 2 남쪽, 3 서쪽
# grid: 0은 육지, 1은 바다, 2는 이미 가 본 육지

n, m = map(int, input().split())
a, b, d = map(int, input().split())
grid = []
answer = 1

for _ in range(n):
  grid.append(list(map(int, input().split())))
grid[a][b] = 2

# 가능 이동경로
da = [-1, 0, 1, 0]
db = [0, 1, 0, -1]
count = 0

while True:
  while count < 4:
    # 현재 방향을 기준으로 왼쪽 방향부터 차례대로 갈 곳을 정한다.
    d = d - 1 if d != 0 else 3 # 방향 돌기
    na, nb = a + da[d], b + db[d]

    # 아직 가보지 않은 칸이 존재하면, 해당 방향으로 한 칸을 전진한다.
    if 0 <= na < n and 0 <= nb < m and grid[na][nb] == 0:
      answer += 1
      grid[na][nb] = 2
      a, b = na, nb
      count = 0
      break

    count += 1

  # 가 보지 않은 칸이 없을 때
  if count == 4:
    # 네 방향 모두 가본 칸이거나 바다로 된 칸인 경우
    # 바라보는 방향을 유지한 채로 한 칸 뒤로 간다
    na, nb = a - da[d], b - db[d]

    # 뒤쪽 칸이 바다인 경우, 움직임을 멈춘다
    if na < 0 or n <= na or nb < 0 or m <= nb:
      break # while True 끝, answer를 print
    elif grid[na][nb] == 1:
      break # while True 끝, answer를 print
    else:
      # 뒤로 가기
      a, b = na, nb
      count = 0

print(answer)
```

* 오타를 내지 않는 것이 생명
 * 너가 낸 오타 - `a, b = na, nb`를 `na, nb = a, b`로 오타냄. `=`를 `==`로 오타냄. 그 외 많음;;;;
* 복잡해지면 함수를 만드는 걸 두려워하지 말 것
* 꼭 while 문을 사용해야 할까? 그냥 if문으로 분기 만들 수도 있는지 생각해 보자.
* 문제를 꼼꼼히 읽으십쇼. '맵의 외곽은 항상 바다로 되어 있다!!' '처음 위치한 칸의 상태는 항상 육지이다!!' 그러면 범위 밖을 빠져나갈 일은 없단 뜻임.

```python
# 모범답안

# 입력 및 맵 생성
# direction: 0 북쪽, 1 동쪽, 2 남쪽, 3 서쪽
# map: 0은 육지, 1은 바다
# visit: 0은 가본 적 없음, 1은 가본 적 있음

n, m = map(int, input().split())
a, b, direction = map(int, input().split())
visit = [[0] * m for _ in range(n)]
visit[a][b] = 1
seamap = []
for _ in range(n):
  seamap.append(list(map(int, input().split())))

# 가능 이동경로
da = [-1, 0, 1, 0]
db = [0, 1, 0, -1]
turn_count = 0
answer = 1

# 좌회전
def turn_left():
  global direction
  direction -= 1
  if direction == -1:
    direction = 3

while True:
  # 현재 방향을 기준으로 왼쪽 방향부터 차례대로 갈 곳을 정한다.
  turn_left()
  na, nb = a + da[direction], b + db[direction]

  # 아직 가보지 않은 칸이 존재하면, 해당 방향으로 한 칸을 전진한다.
  if visit[na][nb] == 0 and seamap[na][nb] == 0:
    answer += 1
    visit[na][nb] = 1
    a, b = na, nb
    turn_count = 0
    continue
  else:
    turn_count += 1

  # 가 보지 않은 칸이 없을 때
  if turn_count == 4:
    # 네 방향 모두 가본 칸이거나 바다로 된 칸인 경우
    # 바라보는 방향을 유지한 채로 한 칸 뒤로 간다
    na, nb = a - da[direction], b - db[direction]

    # 뒤쪽 칸이 바다인 경우, 움직임을 멈춘다
    if seamap[na][nb] == 0:
      a, b = na, nb
    else:
      # 뒤로 가기
      break
    turn_count = 0

print(answer)
```
